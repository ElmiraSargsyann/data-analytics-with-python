## Why Sets are Unordered in Python
### In Python, a set is a collection type that stores unique elements. Unlike lists or tuples, sets do not preserve the order of elements.

### Reason:
- Python sets are implemented as hash tables.
- When you add an element to a set, Python computes a hash value for that element.
- The element is then stored in a position determined by its hash, not by the order you inserted it.
- As a result, when you iterate over a set or convert it back to a list, the original insertion order may be different.

## List Comprehension vs For Loop
### For Loop 
    - Standard way to build lists.
    - More verbose, but often easier to read for beginners.
    - Good for complex logic or multiple steps.
### List Comprehension
    - Shorter and more compact.
    - Works well for simple transformations.
    - Can be less readable for very complex operations.
