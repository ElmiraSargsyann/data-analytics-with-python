```python
import pandas as pd
```

```python
products = pd.read_csv("../data/raw/products.csv")
departments = pd.read_csv("../data/processed/departments_t.csv")
```

```python
products.head()
```

```python
departments.head()
```

```python
df_products_departments = pd.merge(
    products,
    departments,
    on = "department_id",
    how = "left",
    # indicator = True
)
df_products_departments.head()
```

```python
# df_products_departments["_merge"].value_counts()
```

```python
df_products_departments.info()
```

```python
df_products_departments.shape
```

```python
order_train = pd.read_csv("../data/processed/orders_train.csv")
order_products_train = pd.read_csv("../data/raw/order_products_train.csv")
```

```python
order_train.head()
```

```python
order_products_train.head()
```

```python
df_train_transaction = pd.merge(
    order_train,
    order_products_train,
    on = "order_id",
    how = "left",
    # indicator = True
)
df_train_transaction.head()
```

```python
# df_train_transaction["_merge"].value_counts()
```

```python
df_train_transaction.info()
```

```python
df_train_transaction.shape
```

```python
aisles = pd.read_csv("../data/raw/aisles.csv")
aisles.head()
```

```python
df_products_enriched = pd.merge(
    df_products_departments,
    aisles,
    on = "aisle_id",
    how = "left",
    # indicator = True
)
df_products_enriched.head()
```

```python
# df_products_enriched["_merge"].value_counts()
```

```python
df_products_enriched.info()
```

```python
df_products_enriched.shape
```

```python
df_train_analytic = pd.merge(
    df_train_transaction,
    df_products_enriched,
    on = "product_id",
    how = "left",
    # indicator = True
)
df_train_analytic.head()
```

```python
# df_train_analytic["_merge"].value_counts()
```

```python
df_train_analytic.info()
```

```python
df_train_analytic.shape
```

```python
df_customer_states = pd.read_csv("../data/processed/customer_states.csv")
```

```python
df_train_analytics_final = df_train_analytic.merge(
    df_customer_states,
    on = 'user_id',
    how = 'left',
    # indicator=True
)
df_train_analytics_final.head()
```

```python
# df_train_analytics_final["_merge"].value_counts()
```

```python
df_train_analytics_final.drop(columns=['user_id', 'eval_set', 'product_id', 'aisle_id', 'department_id'], inplace = True)
```

```python
print(df_train_analytics_final.shape)
df_train_analytics_final.head()
```

```python
df_train_analytics_final.drop(columns="Unnamed: 0", inplace=True)
```

```python
df_train_analytics_final.to_parquet("../data/processed/instacart.parquet", index = False)
```

```python
df_train_analytics_final.to_csv("../data/processed/instacart.csv", index = False)
```
