```python
# How many items are there in each order?
items_per_order = df_train_transactions.groupby("order_id")["product_id"].count()
items_per_order.head()
```

```python
# Do some days tend to have larger baskets?
avg_items_per_day = df_train_transactions.groupby("order_dow")["product_id"].count() / df_train_transactions.groupby("order_dow")["order_id"].nunique()
avg_items_per_day
```

```python
# At what hour do larger baskets appear?
avg_items_per_hour = df_train_transactions.groupby("order_hour_of_day")["product_id"].count() / df_train_transactions.groupby("order_hour_of_day")["order_id"].nunique()
avg_items_per_hour
```

```python
# Does reorder behavior vary by order timing?
reorder_by_hour = df_train_transactions.groupby("order_hour_of_day")["reordered"].mean()
reorder_by_hour
```
